```python
# Day 4
# While Loop

# while loop is used to repeat a block of code until condition remains TRUE
# while first will check the condition , if the condition is TRUE
# while loop will execute the body
# while will check the condition again
# if the condition is still TRUE while again executes the body

# syntax
# initialization
# while condition:
#   code to be executed
#   incr / decr

i = 1    # initialization            #  i --> 1      again condn will be checked        again condn will be checked
while i<=3:                          #  1<=3 (t)           2<=3 (t)                          3<=3(t)                         4<=3(F)X
  print(i)                           #  print(1)           print(2)                          print(3)
  i = i + 1                          #  i = 2              i = 3                             i = 4
```

```python
# Palindrome Of A Number

# Step 1 : divide the number by 10 and save the remainder
# Step 2 : in order to save the remainder, define one dummy variable and multiply the remainder with 10
# Step 3 : update the number by dividing it by 10
# Step 4 : repeat the above steps until number become ZERO
```

```python
# Palindrome Of A Number

num = 123                       # num = 123          num = 12          num = 1
reverse = 0                        # reverse = 0        reverse = 3       reverse = 32

rem = num % 10                  # rem = 3            rem = 2           rem = 1
reverse = reverse * 10 + rem       # reverse = 3        reverse = 32      reverse = 321
num = num // 10                 # num = 12           num = 1           num = 0
```

```python
num = 121
copy_num = num      # copy_num = 12345
reverse = 0

while num!=0:
  rem = num % 10                # 121!=0 (t)        # 12!=0 (t)       # 1!=0 (t)      # 0!=0 (f)
  reverse = reverse * 10 + rem
  num = num // 10

print(f'Reverse Of A Number : {reverse}')

if copy_num == reverse:     # 12345 == 54321
  print('A Palindrome Number')
else:
  print('Not A Palindrome Number')
```
