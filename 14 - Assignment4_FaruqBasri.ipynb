```python
# PythonSmartApp_UsingAPI

# Final Code
# !pip install wolframalpha==3.0.0
import wolframalpha

# Replace with your Wolfram Alpha App ID
APP_ID = "7VH2K7WUTH"

# Initialize the client
client = wolframalpha.Client(APP_ID)

def query_wolfram(query):
    try:
        res = client.query(query)
        if res.results:
            return next(res.results).text
        else:
            return "No answer found."
    except Exception as e:
        return f"Sorry I cannot answer thatðŸ™ðŸ». Please ask other questionðŸ˜€."

def main():
    print("Python Smart App!")
    print("Ask me anything (type 'exit' to quit).")
    while True:
        user_input = input("\nAsk A Question: ")
        if user_input.lower() in ['exit', 'quit' , 'stop']:
            print("Goodbye!ðŸ‘‹")
            break
        response = query_wolfram(user_input)
        print(f"Answer: {response}")

main()
```

```python
# Guess a number game App

from tkinter import *
import random

w = Tk()
w.title('Guess A Number Game')
w.geometry('600x400')
# background color for window
w.config(bg='#FFD41D')
w.resizable(width=False,height=False)


# main Logic

ranNum = random.randint(1,10)
chance = 4
displayResult = StringVar()

def check_guess():
    global ranNum
    global chance
    userInput = int(user_input.get())
    if chance > 0:
        if userInput == ranNum:
            msg = "Congratulation. You Won! "+str(ranNum)+" is the right number"
            user_input.config(state='disabled')
            guess_button.config(state='disabled')
        elif userInput>ranNum:
            chance = chance - 1
            msg = "Your Guess Is Too High. Think Of A Smaller Number"
            user_input.delete(0,END)
        elif userInput<ranNum:
            chance = chance - 1
            msg = "Your Guess Is Too Low. Think Of A Higher Number"
            user_input.delete(0,END)
    else:
       msg = "Game Over! "+str(chance)+" Attempts Left. The Correct Number Was "+str(ranNum)
       user_input.config(state='disabled')
       guess_button.config(state='disabled')

    displayResult.set(msg)


# Create Widgets
title = Label(w,text='Guess A Number Game',font=('Arial',20),fg='#2B2A2A',bg='#FFA240')
gameInstruction = Label(w,text='Guess The Number Between 1 to 10 in 4 Attempts',
                        font=('Arial',13),fg='#2B2A2A',bg='#FFA240')
user_input = Entry(w,font=('Arial',12))
guess_button = Button(w,text='Guess',font=('Arial',13),fg='white',bg='black',command=check_guess)
exit_button = Button(w,text='Exit Game',font=('Arial',14),fg='white',bg='#b82741',command=w.destroy)
outputLabel = Label(w,font=('Arial',14),fg='black',bg='#FFD41D',textvariable=displayResult)

# Place Widgets
title.place(x=150,y=50)
gameInstruction.place(x=110,y=95)
user_input.place(x=170,y=153)
guess_button.place(x=370,y=147)
exit_button.place(x=230,y=300)
outputLabel.place(x=80,y=220)

w.mainloop()
```

```python
# Sentiment Analyzer App

# !pip install vaderSentiment
# !pip install textblob

from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
sid_obj = SentimentIntensityAnalyzer()

w = Tk()
w.title('Sentiment Analyzer')
w.geometry('600x400')
# background color for window
w.config(bg='#5A9CB5')
w.resizable(width=False,height=False)

# main logic

displayResult = StringVar()

def check_sentiment():
    userInput = user_input.get()
    res = TextBlob(userInput)
    res = sid_obj.polarity_scores(res)

    if res['compound'] >= 0.5:
        displayResult.set('Positive Text')
        user_input.delete(0,END)
    elif res['compound'] <= -0.5:
        displayResult.set('Negative Text')
        user_input.delete(0,END)
    else:
        displayResult.set('Neutral Text')
        user_input.delete(0,END)

# Create Widgets
title = Label(w,text='Sentiment Analyzer \n\nPlease Insert Some Text Below',
             font=('Arial',15),fg='#FACE68',bg='#5A9CB5')
user_input = Entry(w,font=('Arial',14))
button1 = Button(w,text='Click Here To Analyze',font=('Arial',13),fg='#001F3D',
                 bg='#FA6868',command=check_sentiment)
outputLabel = Label(w,font=('Arial',18),fg='#522258',bg='#5A9CB5',
                   textvariable=displayResult)

# Place Widgets
title.place(x=180,y=50)
user_input.place(x=210,y=150)
button1.place(x=235,y=210)
outputLabel.place(x=250,y=300)

w.mainloop()
```

```python
# To-Do List App

from tkinter import *
import tkinter as tk

# Setting title for the App
w = Tk()
w.title("To-Do List")
w.geometry("300x500")
title = Label(w,text='To-Do List',font=('Arial',20),fg='#2B2A2A',bg='#FFEF5F')
title.pack(pady=10)                             # pady used to add vertical spacing

# background color for window
w.config(bg='#3F9AAE')
w.resizable(width=False,height=False)

entry = tk.Entry(w, width=25)                   # width=25 â†’ size of input field
entry.pack(pady=10)

listbox = tk.Listbox(w, width=30, height=15)    # Listbox shows multiple items (tasks)
listbox.pack(pady=10)                           # height=15 means 15 visible rows

def add_task():
    task = entry.get()
    if task != "":
        listbox.insert(tk.END, task)            # listbox.insert() adds the task to the list
        entry.delete(0, tk.END)                 # entry.delete() clears the input box

def delete_task():
    selected = listbox.curselection()           # curselection() finds the selected task
    if selected:
        listbox.delete(selected[0])             # delete() removes it from the list

add_button = tk.Button(w, text="Add Task", command=add_task)
add_button.pack(pady=10)

delete_button = tk.Button(w, text="Delete Task", command=delete_task)
delete_button.pack(pady=10)

w.mainloop()
```
