```python
# Exception handling in Python is a mechanism that allows you to handle runtime errors,
#  or exceptions, in a controlled way, rather than letting your program crash unexpectedly.
# Python provides a way to catch errors, respond to them, and allow your program to continue
# running smoothly even if something goes wrong.
 
# try except finally
# try : keep that block of code inside try , which might give you an error
# except : catch that error
```

```python
# WITHOUT try and except

val1 = int(input('Enter Value 1 : '))
val2 = int(input('Enter Value 2 : '))
res = val1 / val2
print('After Division :',res)
```

```python
# WITHOUT try and except

val1 = int(input('Enter Value 1 : '))
val2 = int(input('Enter Value 2 : '))
res = val1 / val2
print('After Division :',res)
```

```python
# WITH try and except

try:
    val1 = int(input('Enter Value 1 : '))
    val2 = int(input('Enter Value 2 : '))
    res = val1 / val2
    print('After Division :',res)
except:
    print('Error')
    print('Cannot Divide By Zero or Please Provide The Integer Input')
```

```python
# Let's try to get the Exact Root Cause Of THe Error
# Using Parent Exception Class

try:
    val1 = int(input('Enter Value 1 : '))
    val2 = int(input('Enter Value 2 : '))
    res = val1 / val2
    print('After Division :',res)
except Exception as e:
    print('Error :',e)
```

```python
try:
    list1 = [10,20,30,40]
    print(list1[1])
    print(list1[12])
except Exception as e:
    print('Error :',e)
print('rest of the code')
```

```python
# Using Exact Exception Class

try:
    val1 = int(input('Enter Value 1 : '))
    val2 = int(input('Enter Value 2 : '))
    res = val1 / val2
    print('After Division :',res)
except ArithmeticError as e:
    print('Error :',e)
```

```python
try:
    val1 = int(input('Enter Value 1 : '))
    val2 = int(input('Enter Value 2 : '))
    res = val1 / val2
    print('After Division :',res)
except (ZeroDivisionError,ValueError) as e:
    print('Error :',e)
```

```python
# finally
# finallt is a block of code which will Execute Always

# CASE 1 : WHEN THERE IS NO EXCEPTION

try:
    val1 = int(input('Enter Value 1 : '))
    val2 = int(input('Enter Value 2 : '))
    res = val1 / val2
    print('After Division :',res)
except Exception as e:
    print('Error :',e)
finally:
    print('I will execute always')
```

```python
# CASE 2 : WHEN THERE IS AN EXCEPTION

try:
    val1 = int(input('Enter Value 1 : '))
    val2 = int(input('Enter Value 2 : '))
    res = val1 / val2
    print('After Division :',res)
except Exception as e:
    print('Error :',e)
finally:
    print('I will execute always')

print('Rest Of The Code')
```

```python
# CASE 3 : WHEN THERE IS AN EXCEPTION BUT NOT HANDELLED PROPERLY

try:
    val1 = int(input('Enter Value 1 : '))
    val2 = int(input('Enter Value 2 : '))
    res = val1 / val2
    print('After Division :',res)
except KeyError as e:
    print('Error :',e)
finally:
    print('I will execute always')

print('Rest Of The Code')
```
